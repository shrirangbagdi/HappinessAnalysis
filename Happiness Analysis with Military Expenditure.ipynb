```python
import pandas as pd 
import matplotlib as plt 
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
military_df = pd.read_csv("Desktop/Military Expenditure.csv")
military_df
```

```python
happiness_df = pd.read_csv("desktop/2019.csv")
happiness_df
```

```python
military_df = military_df.groupby("Name").mean()
```

```python
for country in happiness_df["Country or region"]:
    if country in military_df.index:
        value = military_df["2018"][country]
        happiness_df.loc[happiness_df["Country or region"] == country, "Military Spending"] = value
```

```python
happiness_df
```

```python
happiness_df.dropna(inplace=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = happiness_df[["GDP per capita", "Social support", "Healthy life expectancy", "Freedom to make life choices", "Generosity", "Perceptions of corruption", "Military Spending" ]]
y = happiness_df["Score"]
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
```

```python
plt.pyplot.scatter(y_test,predictions)
```

```python

from sklearn import metrics

print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

```python
sns.distplot((y_test-predictions),bins=50);
```

```python

coeffecients = pd.DataFrame(model.coef_,X.columns)
coeffecients.columns = ['Coeffecient']
coeffecients
```
